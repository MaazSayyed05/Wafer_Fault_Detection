```python
# Get columns to drop
def get_cols_with_zero_std_dev(df: pd.DataFrame):
    """Returns a list of COLUMNS names who are having zero std dev."""
    cols_to_drop = []
    num_cols = [col for col in df.columns if df[col].dtype != 'O']
    for col in num_cols:
        if df[col].std() == 0:
            cols_to_drop.append(col)
    return cols_to_drop



def get_reduntant_cols(df: pd.DataFrame, missing_thresh= 0.7):
    """"
    Returns a list of COLUMNS having missing values more than certain thresh."""

    cols_missing_ratios = df.isna().sum().div(df.shape[0])
    cols_to_drop = list(cols_missing_ratios[cols_missing_ratios > missing_thresh].index)
    return cols_to_drop
```

```python
# Drop columns with missing more than 70%
cols_to_drop_1 = get_reduntant_cols(wafers_train,missing_thresh=0.7)
cols_to_drop_1
```

```python
# Drop columns with 0 std. dev.
cols_to_drop_2 = get_cols_with_zero_std_dev(wafers_train)
cols_to_drop_2
```

```python
cols_to_drop = cols_to_drop_1 + cols_to_drop_2
cols_to_drop
```

```python
# Separate Features and Labels Out
X,y = wafers_train.drop(cols_to_drop,axis=1),wafers_train[['Good/Bad']]
```
